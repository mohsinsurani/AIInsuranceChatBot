```python
from PyPDF2 import PdfReader
from langchain.text_splitter import (
    CharacterTextSplitter,RecursiveCharacterTextSplitter
)
from langchain.embeddings.huggingface import HuggingFaceInstructEmbeddings
from langchain_community.vectorstores import FAISS
import pickle
import os.path
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma
from langchain.document_loaders import UnstructuredMarkdownLoader, UnstructuredURLLoader
import torch
from langchain import PromptTemplate
from langchain.chains import LLMChain, SimpleSequentialChain, RetrievalQA, ConversationalRetrievalChain
```

```python
def get_pdf_text(pdf_docs):
    text = ""
    for pdf in pdf_docs:
        pdf_reader = PdfReader(pdf)
        for page in pdf_reader.pages:
            if page is not None:
                page_text = page.extract_text()
                if page_text is not None:
                    text += page_text
    return text

def get_text_chunks(raw_text):
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=1024, chunk_overlap=64)
    texts_chunks = text_splitter.split_text(raw_text)
    return texts_chunks
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"

embeddings = HuggingFaceEmbeddings(
model_name="thenlper/gte-large",
model_kwargs={"device": device},
encode_kwargs={"normalize_embeddings": True},
)
```

```python
pdf_docs = open("healthinsurance.pdf", "rb")
raw_text = get_pdf_text(pdf_docs=[pdf_docs])
text_chunks = get_text_chunks(raw_text)
```

```python
# db = Chroma.from_documents(text_chunks, embeddings, persist_directory="db")
db = Chroma.from_texts(text_chunks, embeddings, persist_directory="db")
```
